```python
def document_frequencies():    
    for term in dictionary_wf:
        document_frequency[term] = len(postings[term])
```

```python
def tfidf_weights(term,id):
    if id in postings[term]["DocID"].values:
        for j in index[term]:
            return j[1]*idf(term)
    else:
        return 0.0
    
def idf(term):
    if term in dictionary_wf:
        return math.log(N/document_frequency[term],2)
    else:
        return 0.0
```

```python
def vector_norm():
    for id in all_docs:
        c = 0
        for term in dictionary_wf:
            c += index[term][1]**2
        norm[id] = math.sqrt(c)  
# rever
```

```python
def cosine_tfidf_norm():
    for id in all_docs:
        c = 0
        for term in dictionary_wf:
            c += tfidf_weights(term,id)**2
        norm_tfidf[id] = math.sqrt(c)
```

```python
def similarity_cosine(query,id):
    similarity = 0.0
    for term in query:
        if term in dictionary_wf:
            similarity += index[term][1]*document_frequency[term]
    similarity = similarity / norm[id]
    return similarity
# rever
```

```python
def similarity_cosine_tfidf(query,id):
    similarity = 0.0
    for term in query:
        if term in dictionary_wf:
            similarity += tfidf_weights(term,id)*idf(term)
    similarity = similarity / norm_tfidf[id]
    return similarity
```
